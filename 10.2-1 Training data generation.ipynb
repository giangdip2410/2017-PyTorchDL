```python
import torch
import matplotlib.pyplot as plt
```

```python
Sample_size = 100
```

# Generate 2d data using two independent normal distribution

```python
n_data = torch.ones(Sample_size, 1) # [1, 1, 1, ...,1]
```

```python
x11 = torch.normal(-3*n_data, 2)
x12 = torch.normal(-3*n_data, 2)
x21 = torch.normal(3*n_data, 2)
x22 = torch.normal(3*n_data, 2)
x31 = torch.normal(0*n_data, 1)
x32 = torch.normal(0*n_data, 1)
x41 = torch.normal(-2*n_data, 1)
x42 = torch.normal(n_data, 1)
x51 = torch.normal(0*n_data, 1)
x52 = torch.normal(2*n_data, 1)
x61 = torch.normal(0*n_data, 1)
x62 = torch.normal(-2*n_data, 1)
```

```python
x1=torch.cat((x11, x12), 1)
x2=torch.cat((x21, x22), 1)
x3=torch.cat((x31, x32), 1)
x4=torch.cat((x41, x42), 1)
x5=torch.cat((x51, x52), 1)
x6=torch.cat((x61, x62), 1)
```

```python
# visualize x1, x2, ..., x6
fig = plt.figure(figsize=(18, 9)) 

a=fig.add_subplot(2,3, 1) # we have a 2 rows * 3 columns subplots
plt.axis([-8, 8, -8, 8]) #minX, maxX, minY, maxY
plt.scatter(x1.numpy()[:, 0], x1.numpy()[:, 1], c='blue', s=20)
plt.title('X1')

a=fig.add_subplot(2,3, 2) 
plt.axis([-8, 8, -8, 8]) #minX, maxX, minY, maxY
plt.scatter(x2.numpy()[:, 0], x2.numpy()[:, 1], c='blue', s=20)
plt.title('X2')

a=fig.add_subplot(2,3, 3) 
plt.axis([-8, 8, -8, 8]) #minX, maxX, minY, maxY
plt.scatter(x3.numpy()[:, 0], x3.numpy()[:, 1], c='blue', s=20)
plt.title('X3')

a=fig.add_subplot(2,3, 4) 
plt.axis([-8, 8, -8, 8]) #minX, maxX, minY, maxY
plt.scatter(x4.numpy()[:, 0], x4.numpy()[:, 1], c='blue', s=20)
plt.title('X4')

a=fig.add_subplot(2,3, 5) 
plt.axis([-8, 8, -8, 8]) #minX, maxX, minY, maxY
plt.scatter(x5.numpy()[:, 0], x5.numpy()[:, 1], c='blue', s=20)
plt.title('X5')

a=fig.add_subplot(2,3, 6) 
plt.axis([-8, 8, -8, 8]) #minX, maxX, minY, maxY
plt.scatter(x6.numpy()[:, 0], x6.numpy()[:, 1], c='blue', s=20)
plt.title('X6')

plt.show()
```

```python
y1 = torch.ones(100)
y2 = 2*y1
y3 = 3*y1
y4 = 4*y1
y5 = 5*y1
y6 = 6*y1
```

# x = x1+ x2+ ...+ x6

```python
x = torch.cat((x1, x2), 0)
x = torch.cat((x, x3), 0)
x = torch.cat((x, x4), 0)
x = torch.cat((x, x5), 0)
x = torch.cat((x, x6), 0).type(torch.FloatTensor)
```

```python
y = torch.cat((y1, y2), 0)
y = torch.cat((y, y3), 0)
y = torch.cat((y, y4), 0)
y = torch.cat((y, y5), 0)
y = torch.cat((y, y6), 0).type(torch.LongTensor)
```
